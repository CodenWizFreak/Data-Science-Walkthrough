```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
def gradient_descent(x,y): #gradient descent makes small adjustments to reduce errors in predictions. e.g. climbing down the steepest part of a hill to reach the bottom quickly.
    m_curr = b_curr = 0
    rate = 0.084 #learning rate is greater than 10^-6 and lesser than 1. e.g. you're moving downhill and how big steps you should take to reach the bottom but you should not surpass the bottom.
    n = len(x)
    plt.scatter(x,y,color='red',marker='+',linewidth=5)
    for i in range(10000):#how many baby steps you wanna take to the bottom of the hill
        y_predicted = m_curr * x + b_curr
        cost=(1/n)*sum([val**2 for val in (y-y_predicted)]) #e.g. the more a dart is further away from the bull's eye, the greater the cost function. Need to keep it reducing or decreasing to keep the darts hit closer to the bull's eye.
        #cost function if increases means that the values might be wrong. you're going away from the bull's eye instead of going towards it.
        print (f"m_curr={m_curr}; b_curr={b_curr}; cost={cost}; i={i}")
        plt.plot(x,y_predicted,color='green')
        md = -(2/n)*sum(x*(y-y_predicted))
        yd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * yd
        
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```
