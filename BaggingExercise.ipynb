```python
df2 = df1[df1.RestingBP<=(df1.RestingBP.mean()+3*df1.RestingBP.std())]
df2.shape
```

```python
df2[df2.Oldpeak>(df2.Oldpeak.mean()+3*df2.Oldpeak.std())]
```

```python
df3 = df2[df2.Oldpeak<=(df2.Oldpeak.mean()+3*df2.Oldpeak.std())]
df3.shape
```

```python
df3.RestingECG.unique()
```

```python
df3.ST_Slope.unique()
```

```python
df4=df3.copy()
df4.RestingECG.replace({'Normal':1,'ST':2,'LVH':3},inplace=True)
df4.ST_Slope.replace({'Up':1,'Flat':2,'Down':3},inplace=True)
df4.head()
```

```python
df5=pd.get_dummies(df4,drop_first=True).astype(int)
df5.head()
```

```python
X = df5.drop("HeartDisease",axis='columns')
y = df5.HeartDisease
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scale=scaler.fit_transform(X)
X_scale
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scale, y, test_size=0.2, random_state=30)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
scores=cross_val_score(SVC(),X,y,cv=5)
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier
bag_model = BaggingClassifier(estimator=SVC(), n_estimators=100, max_samples=0.8, oob_score=True,random_state=0)
scores = cross_val_score(bag_model, X, y, cv=5)
scores.mean()
```

```python
from sklearn.tree import DecisionTreeClassifier
scores = cross_val_score(DecisionTreeClassifier(random_state=0), X, y, cv=5)
scores.mean()
```

```python
bag_model = BaggingClassifier(estimator=DecisionTreeClassifier(random_state=0),n_estimators=100,max_samples=0.9,oob_score=True,random_state=0)
scores = cross_val_score(bag_model, X, y, cv=5)
scores.mean()
```
