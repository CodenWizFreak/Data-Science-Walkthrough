```python
# This is Feature Engineering using percentile
import pandas as pd
```

```python
df=pd.read_csv('C:/Users/dasgu/Desktop/MyImp/Code/aiml/heights.csv')
df.head()
```

```python
max_threshold=df.height.quantile(0.95) #We are taking note of the upper most outliers
max_threshold
```

```python
df[df.height>max_threshold] #This row is the outlier one
```

```python
min_threshold=df.height.quantile(0.05) #We are taking note of the lower most outliers
min_threshold
```

```python
df[df.height<min_threshold] #This row is the outlier one
```

```python
df[(df.height<max_threshold)&(df.height>min_threshold)] #removing the outliers
```
