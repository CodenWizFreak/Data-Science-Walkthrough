```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('C:/Users/dasgu/Desktop/MyImp/Code/aiml/canada_per_capita_income.csv')
df.head(3)
```

```python
reg=linear_model.LinearRegression() #Linear regression has one dependent and independent variable, while multivariable regression has two or more.
reg.fit(df[['year']],df['per capita income (US$)']) #trains the model by finding the best fitting line or curve (income = a+b*year).
#here income is the dependant variable and year is the independant variable.
```

```python
%matplotlib inline
plt.xlabel('year',fontsize=20)#keep 'to be entered' data in x axis
plt.ylabel('per capita income (US$)',fontsize=20)#keep 'to be predicted' data in y axis
plt.scatter(df['year'],df['per capita income (US$)'],color='red')
plt.plot(df['year'],reg.predict(df[['year']]),color='green')
```
